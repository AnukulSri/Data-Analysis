```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
df.head(4)
```

```python
dff = pd.read_csv('ratings.csv')
dff
```

# removing the column which are not need from both the dataframe

```python
cols_to_drop = ['genres']
df.drop(columns=cols_to_drop,axis=0,inplace= True)
```

```python
dff.drop(columns='timestamp',axis=0,inplace=True)
```

# merge both the dataframe

```python
n_df = pd.merge(df,dff,on='movieId')
n_df
```

```python
combine_movie_rating = n_df.dropna(axis = 0, subset = ['title'])
movie_ratingCount = (combine_movie_rating.
     groupby(by = ['title'])['rating'].
     count().
     reset_index().
     rename(columns = {'rating': 'totalRatingCount'})
     [['title', 'totalRatingCount']]
    )
movie_ratingCount.head()
```

```python
n_df = pd.merge(n_df,movie_ratingCount,on='title')
```

```python
n_df
```

```python
popular_movie = n_df.query('totalRatingCount >= 50')
n_df.head(5)
```

# create pivot table

```python
moviemat = n_df.pivot_table(index='userId',columns='title',values='rating').fillna(0)
moviemat.head()
```
